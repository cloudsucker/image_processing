```python
import os
import cv2
from google.colab.patches import cv2_imshow
```

#### **Медианный фильтр**


```python
IMAGES_PATH = 'images/'
image_filenames = os.listdir(IMAGES_PATH)
images_size = (512, 512)

class ImageNotFound(Exception):
    def __init__(self, message: str):
        self.message = message

row_result = None
processed_result = None

for image_filename in image_filenames:
    image_path = IMAGES_PATH + image_filename
    image = cv2.resize(cv2.imread(image_path), images_size)

    if image is None:
        raise ImageNotFound(f"Ошибка: изображение '{image_path}' не найдено.")

    processed_image = cv2.resize(cv2.medianBlur(image, 3), images_size)

    if (row_result is not None) and (processed_image is not None):
        row_result = cv2.vconcat([row_result, image])
        processed_result = cv2.vconcat([processed_result, processed_image])
    else:
        row_result = image
        processed_result = processed_image

result = cv2.hconcat([row_result, processed_result])
cv2_imshow(result)
cv2.imwrite('first_task_median_result.png', result)
```

**Описание результата:**

Медианный фильтр отлично показал себя на этой выборке. Он не только сгладил шумы, но и убрал их явно проявленные фрагменты. К примеру на третьем изображении мы смогли хоть как-то разглядеть изображение, которое казалось, никак не исправить.

#### **Средний фильтр**


```python
import numpy as np
```

```python
IMAGES_PATH = 'images/'
image_filenames = os.listdir(IMAGES_PATH)
images_size = (512, 512)

row_result = None
processed_result = None

for image_filename in image_filenames:
    image_path = IMAGES_PATH + image_filename
    image = cv2.resize(cv2.imread(image_path), images_size)

    if image is None:
        raise ImageNotFound(f"Ошибка: изображение '{image_path}' не найдено.")

    kernel = np.ones((3, 3), np.float32) / 9
    processed_image = cv2.resize(cv2.filter2D(image, -1, kernel), images_size)

    if (row_result is not None) and (processed_image is not None):
        row_result = cv2.vconcat([row_result, image])
        processed_result = cv2.vconcat([processed_result, processed_image])
    else:
        row_result = image
        processed_result = processed_image

result = cv2.hconcat([row_result, processed_result])
cv2_imshow(result)
cv2.imwrite('first_task_average_result.png', result)
```

**Описание результата:**

Средний фильтр на данной выборке показал худший в сравнении с медианным фильтром результат. Он сглаживает изображение, но не убирает сами шумы, лишь уменьшает их силу.

```python
IMAGES_PATH = "/content/unzipped_images/IMAGE/Image_1/"
MAIN_IMAGE_FILENAME = 'IMAGE.jpg'
image = cv2.imread(IMAGES_PATH + MAIN_IMAGE_FILENAME)

if image is None:
    raise Exception(f"Ошибка: изображение '{IMAGES_PATH + MAIN_IMAGE_FILENAME}' не найдено.")

image_with_boxes = image.copy()
fragment_images = None

for i in range(9):
    fragment = cv2.imread(IMAGES_PATH + f'Et_{i+1}.png')

    if fragment is None:
        raise Exception(f"Ошибка: фрагмент '{IMAGES_PATH + f'Et_{i+1}.png'}' не найден.")

    # Получаем размеры фрагмента
    (fragmentHeight, fragmentWidth) = fragment.shape[:2]

    # Осуществляем поиск фрагмента на изображении
    result = cv2.matchTemplate(image, fragment, cv2.TM_CCOEFF)
    (_, _, minLoc, maxLoc) = cv2.minMaxLoc(result)

    # Выделяем искомый фрагмент
    topLeft = maxLoc
    botRight = (topLeft[0] + fragmentWidth, topLeft[1] + fragmentHeight)

    cv2.rectangle(image_with_boxes, topLeft, botRight, (0, 255, 0), 3)

    fragment = cv2.resize(fragment, (100, 100))
    if fragment_images is not None:
        fragment_images = cv2.hconcat([fragment_images, fragment])
    else:
        fragment_images = fragment

# image_with_boxes = cv2.resize(image_with_boxes, (1000, 800))

cv2_imshow(image_with_boxes)
cv2_imshow(fragment_images)

cv2.imwrite('second_task_result.png', image_with_boxes)
```
