```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def convert_to_gray(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return gray

def apply_gaussian_blur(gray_image, kernel_size=(15, 15)):
    blurred = cv2.GaussianBlur(gray_image, kernel_size, 0)
    return blurred

def apply_canny_edge_detection(blurred_image, threshold1=100, threshold2=200):
    edges = cv2.Canny(blurred_image, threshold1, threshold2)
    return edges
```

```python
def detect_circles(edges_image, dp=1.2, minDist=50, param1=50, param2=30, minRadius=15, maxRadius=50):
    circles = cv2.HoughCircles(edges_image, cv2.HOUGH_GRADIENT, dp, minDist, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)
    return circles

def classify_and_sum_coins(circles, image):
    coin_sizes = {
        "0.01 RUB": (5, 7),
        "0.05 RUB": (7, 9),
        "0.10 RUB": (9, 10),
        "0.50 RUB": (5, 15),
        "2 RUB": (35, 40),
        "10 RUB": (40, 45),
        "5 RUB": (45, 50)
    }

    total_sum = 0

    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")

        for (x, y, r) in circles:
            for coin, (min_radius, max_radius) in coin_sizes.items():
                if min_radius <= r <= max_radius:
                    cv2.putText(image, coin, (x - 10, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1, cv2.LINE_AA)
                    total_sum += int(coin.split()[0])

            cv2.circle(image, (x, y), r, (0, 255, 0), 4)

    return total_sum

def process_image(image_path):
    image = cv2.imread(image_path)

    gray_image = convert_to_gray(image)
    blurred_image = apply_gaussian_blur(gray_image)
    edges_image = apply_canny_edge_detection(blurred_image)

    circles = detect_circles(edges_image)
    total_sum = classify_and_sum_coins(circles, image)

    print(f"Общая сумма: {total_sum} рублей")
    cv2_imshow(image)

image_path = "rubles.jpg"
process_image(image_path)
```
